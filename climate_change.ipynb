```python
import pandas as pd 
```

```python


# converted date to index
wd = pd.read_csv('climate_change.csv', parse_dates=['date'], index_col='date')
 
wd


```

```python
# replace the missing values in co2 with the median

wd["co2"] = wd["co2"].fillna(wd['co2'].median())
wd
```

```python
#replace temperature negative number with 0
wd['relative_temp'] = wd['relative_temp'].clip(lower=0)
wd
```
